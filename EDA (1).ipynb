#### make function to handle outliers of columns with IQR Method by capping

```python
def iqr_technique(DFcolumn):
  Q1 = np.percentile(DFcolumn, 25)
  Q3 = np.percentile(DFcolumn, 75)
  IQR = Q3 - Q1
  lower_range = Q1 - (1.5 * IQR)
  upper_range = Q3 + (1.5 * IQR)
  return lower_range, upper_range
```

```python
lower_range, upper_range = iqr_technique(housing_data['SalePrice'])
housing_data = housing_data[(housing_data.SalePrice > lower_range) & (housing_data.SalePrice<upper_range)]
```

```python
# Check that the outliers are removed.
sns.boxplot(x = housing_data['SalePrice'])
print(housing_data.shape)
```

### 3. Average price of Each Enighbourhood group using Point Plot

```python
housing_data.head()
```

```python
# Group the housing dataset by Neighborhood and calculate the mean of each group
#grouped = housing_data.groupby("Neighborhood").mean()
grouped = housing_data.groupby("Neighborhood")["SalePrice"].mean()
# Reset the indes of the grouped data so that the Neighborhood becomes a column
Neighborhood_avg_SalePrice = grouped.reset_index()

#Rename the price to 'avg_SalePrice'
Neighborhood_avg_SalePrice = round(Neighborhood_avg_SalePrice.rename(columns={"SalePrice": 'avg_SalePrice'}), 50)


# Select only the 'Neighborhood', and the avg_SalePrice
Neighborhood_avg_SalePrice[['Neighborhood', 'avg_SalePrice']].head()
```

```python
# Import mean from the statistics module
from statistics import mean

# Create the pointplot
sns.pointplot(x = 'Neighborhood', y = 'SalePrice', data=housing_data, estimator = np.mean)

# Add axis and labels
plt.xlabel("Neighborhood")

plt.ylabel('Average SalePrice')

plt.title('Average SalePrice by the Neighborhood')
```
